```python
# Represnt the sales on a bar chart , and also show the market share for each location using a pie chart
```

```python
location_list=sales.groupby('Location')
```

```python
location=[x for x,y in location_list]
```

```python
location
```

```python
import matplotlib.pyplot as plot # import plot
```

```python
plot.bar(location,sales.groupby('Location').sum()['Total']) # to represent bar
```

```python
plot.pie(sales.groupby('Location').sum()['Total'],labels=location,autopct='%1.f%%')
```
